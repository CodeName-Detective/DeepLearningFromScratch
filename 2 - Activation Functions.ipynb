### Implementation as CLASS

```python
import numpy as np
```

```python
inputs = [[1,2,3,2.5],
         [2.0,5.0,1.0,2.0],
         [-1.5,2.7,3.3,-0.8]]
```

```python
class DenseLayer:
    def __init__(self,n_inputs,n_neurons):
        self.weights = 0.10 * np.random.randn(n_inputs,n_neurons)
        self.biases = np.zeros((1,n_neurons))
    def forward(self,inputs):
        self.output = np.dot(inputs,self.weights)+self.biases
```

```python
class Activation_Relu:
    def forward(self,inputs):
        self.output = np.maximum(inputs,0) 
```

#### Creation of Data

```python
np.random.seed(17)

def create_data(points, classes):
    X = np.zeros((points*classes, 2))
    y = np.zeros(points*classes, dtype='uint8')
    for class_number in range(classes):
        ix = range(points*class_number, points*(class_number+1))
        r = np.linspace(0.0,1,points) #Radius
        t = np.linspace(class_number*4,(class_number+1)*4, points) + np.random.randn(points)*0.2
        X[ix] = np.c_[r*np.sin(t*2.5),r*np.cos(t*2.5)]
        y[ix] = class_number
    return X,y
```

```python
X,y = create_data(100,3)
```

```python
layer1 = DenseLayer(2,50)
activation1 = Activation_Relu()
```

```python
layer1.forward(X)
```

```python
layer1.output
```

```python
activation1.forward(layer1.output)
```

```python
activation1.output
```

## Softmax (Exponentiation + Normalization)

```python
layer_outputs = [4.8,1.21,2.385]
```

```python
import math
E = math.e
```

```python
#Exponentiation
exp_values = []
for output in layer_outputs:
    exp_values.append((E**output))
```

```python
exp_values
```

```python
#Normalization
norm_base = sum(exp_values)
norm_values = []
for value in exp_values:
    norm_values.append(value/norm_base)
```

```python
norm_values
```

```python
sum(norm_values)
```

### With Numpy

```python
exp_values = np.exp(layer_outputs)
norm_values = exp_values / np.sum(exp_values)
norm_values
```

```python
np.sum(norm_values)
```

### Exponential Overflow

```python
np.exp(1000)
```

* To prevent overflow during exponentiation v = u - max(u)
* Input will be from -inf to 0
* which makes range of possibilities after exponentiation between 0 and 1

```python
layer_outputs = [[4.8,1.21,2.385],
                [8.9,-1.81,0.2],
                [1.41,1.051,0.026]]
```

```python
max_values = np.max(layer_outputs,axis=1,keepdims=True)
```

```python
layer_outputs -= max_values
```

```python
layer_outputs
```

```python
exp_values = np.exp(layer_outputs)
exp_values
```

```python
norm_values = exp_values / np.sum(exp_values,axis=1,keepdims=True)
norm_values
```

```python
np.sum(norm_values,axis=1,keepdims=True)
```

```python
class Activation_Softmax:
    def forward(self,inputs):
        exp_values = np.exp(inputs-np.max(inputs,axis=1,keepdims=True))
        norm_values = exp_values / np.sum(exp_values,axis=1,keepdims=True)
        self.output = norm_values
```

```python
X,y = create_data(100,3)
```

```python
layer1 = DenseLayer(2,3)
activation1 = Activation_Relu()
layer2 = DenseLayer(3,3)
activation2 = Activation_Softmax()
```

```python
layer1.forward(X)
activation1.forward(layer1.output)

layer2.forward(activation1.output)
activation2.forward(layer2.output)
```
